```python
################################### Correct Code for UTube Video Download with Progress Bar #########################
#!pip install tkinter
#!pip install pytube
#####################################################################################################################

from tkinter import filedialog
from tkinter import *
from pytube import YouTube
import time
from threading import *
from tkinter import ttk

def progress(streams, chunk: bytes, bytes_remaining: int):
    global video
    contentsize = video.filesize
    OldSize = contentsize - bytes_remaining
        
    OldMax = contentsize
    OldMin = 0
    NewMax = 100
    NewMin = 0
    OldRange = (OldMax - OldMin)  
    NewRange = (NewMax - NewMin)  
    NewSize = (((OldSize - OldMin) * NewRange) / OldRange) + NewMin
    
    pbar = ((OldSize/OldMax)*100)
    bar['value'] = pbar
    download = pbar
    percent.set(str(int((download/NewMax)*100))+"%")
    text.set(str(int(OldSize))+"/"+str(OldMax)+" Bytes completed")
    root.update_idletasks()
       
def browse_folder():
    global folder_path, filepath
    filepath = filedialog.askdirectory()
    folder_path.set(filepath)

def Downloader():     
    global filepath, video
    url =YouTube(str(link.get()),on_progress_callback=progress)
    video = url.streams.first()
    st = time.time()
    video.download(filepath)
    tt = time.time() - st
    titext = 'Download Time: {:.4f} Sec.'.format(tt)
    Label(root, text = 'DOWNLOADED', font = 'arial 15').place(x= 180 , y = 210)
    Label(root, text = titext, font ='arial 10 bold', fg='red').place(x= 180 , y = 250)
    link.set("")

def startDownloadThread():
    thread=Thread(target=Downloader)
    thread.start()    
    
root = Tk()
root.geometry('500x300')
root.resizable(0,0)
root.title("RATech-UTube Video Downloader")

Label(root, text = 'Utube Video Downloader', font ='arial 20 bold').pack()
bar = ttk.Progressbar(root,orient=HORIZONTAL,length=100,mode='determinate')
bar.pack(pady=5)

link = StringVar()
folder_path = StringVar()
filepath = "./"
percent = StringVar()
text = StringVar()

Label(root, text = 'Paste Link Here-', font = 'arial 15 bold').place(x= 160 , y = 60)
link_enter = Entry(root, width = 70,textvariable = link).place(x = 32, y = 90)
percentLabel = Label(root, textvariable = percent, font = 'arial 10 bold', fg = 'blue').place(x= 310 , y = 40)
taskLabel = Label(root, textvariable = text, font = 'arial 10 bold', fg = 'blue').pack(pady=50)


Label(root, text = 'Download Time: 0 Sec.', font ='arial 10 bold', fg='red').place(x= 180 , y = 250)    
Button(root,text = 'BROWSE', font = 'arial 15 bold', bg = 'yellow', padx = 2, command = browse_folder).place(x=120 ,y = 150)
Button(root,text = 'DOWNLOAD', font = 'arial 15 bold', bg = 'pale green', padx = 2, command = startDownloadThread).place(x=250 ,y = 150)

root.mainloop()
```
